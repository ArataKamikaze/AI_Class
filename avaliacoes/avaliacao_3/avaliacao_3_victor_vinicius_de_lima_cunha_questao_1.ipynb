```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
title = ['class', 'Alcohol', 'Malicacid', 'Ash', 'Alcalinity_of_ash', 'Magnesium', 'Total_phenols', 'Flavanoids', 'Nonflavanoid_phenols', 'Proanthocyanins', 'Color_intensity', 'Hue', '0D280_0D315_of_diluted_wines', 'Proline']
raw_data = pd.read_csv("wine.data", names=title, index_col= False)
raw_data
```

```python
sns.pairplot(data=raw_data)
```

```python
#importando as bibliotecas dos modelos
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
#import the splitter
from sklearn.model_selection import train_test_split
```

```python
#import the scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#import the metrics
from sklearn.metrics import confusion_matrix, classification_report
```

# KNN

```python
knn_scaled_data = scaler.fit_transform(raw_data.drop('class', axis = 1))
knn_feats = pd.DataFrame(knn_scaled_data,columns=raw_data.drop('class', axis=1).columns)
knn_x = knn_feats
knn_y = raw_data['class']
```

```python
knn_X_train, knn_X_test, knn_y_train, knn_y_test = train_test_split(knn_x, knn_y, test_size=0.3, random_state=42)
```

```python
error_rate = []

# Will take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(knn_X_train,knn_y_train)
    pred_i = knn.predict(knn_X_test)
    error_rate.append(np.mean(pred_i != knn_y_test))
```

```python
plt.figure(figsize=(20,4))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=1)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors=15)
knn.fit(knn_X_train,knn_y_train)
pred = knn.predict(knn_X_test)
print('\n')
print(confusion_matrix(knn_y_test,pred))
print('\n')
print(classification_report(knn_y_test,pred))
```

# Logistical Regression

```python
lrx = raw_data.drop('class', axis = 1)
lry = raw_data['class']
lry = lry.replace({2: 0, 3: 0})
```

```python
X_train, X_test, y_train, y_test = train_test_split(lrx, lry, test_size=0.3, random_state=101)
```

```python
logmodel = LogisticRegression(max_iter=1000)
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
print('\n')
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```
